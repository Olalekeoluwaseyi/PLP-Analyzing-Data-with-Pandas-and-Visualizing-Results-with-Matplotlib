```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("iris.data", header=None)
dataset.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
dataset.head()
```

```python
dataset.info()
```

```python
print(dataset.isnull().sum())
```

Task 2: Basic Data **Analysis**

```python
dataset.describe()
```

```python
mean_of_group_data = dataset.groupby("class")['petal length'].mean()
print(mean_of_group_data)
```

**Observation**

1. Petal length is a strong feature for distinguishing between iris classes.

2. The clear separation in the means suggests that petal length can be used for classification, especially to separate Setosa from the others.

3. There is low overlap in petal lengths between the classes, particularly between Setosa and Virginica.

Task 3: Data **Visualization**

```python
import matplotlib.pyplot as plt
```

```python
# Simulate a date range for time series
dataset['date'] = pd.date_range(start='1998-01-01', periods=len(dataset), freq='D')

dataset_time = dataset.groupby('date')['petal length'].mean()

plt.figure(figsize=(10, 5))
plt.plot(dataset_time.index, dataset_time.values, color='green')
plt.title('Average Petal Length Over Time')
plt.xlabel('Date')
plt.ylabel('Petal Length')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**2. Bar Chart  Average Petal Length per Species**

```python
import seaborn as sns

plt.figure(figsize=(7, 5))
sns.barplot(x='class', y='petal length', data= dataset, palette='Set2')
plt.title('Average Petal Length by Species')
plt.xlabel('Iris Species')
plt.ylabel('Average Petal Length')
plt.tight_layout()
plt.show()
```

**Histogram  Distribution of Sepal Width**

```python
plt.figure(figsize=(7, 5))
plt.hist(dataset['sepal width'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**4. Scatter Plot  Sepal Length vs. Petal Length**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length', y='petal length', hue='class', data=dataset, palette='Dark2')
plt.title('Sepal Length vs. Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```
